```python
# Let's check for the 10 most popular songs on spotify ;)

most_popular = tracks.query('popularity > 90',inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
#Convert song duration from ms to seconds

tracks["duration"] = tracks["duration_ms"].apply(lambda x: round(x/1000))
tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
#Heatmap between variables 



corr_df = tracks.drop(["key", "mode", "explicit"], axis = 1).corr(method = "pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt=".1g", vmin = -1, vmax = 1, center = 0, cmap = "crest", linewidths = 1, linecolor = 'Black')
heatmap.set_title("Correlation Heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df = tracks.sample(int(0.004 *len(tracks)))
print(len(sample_df))
```

```python
#Inspecting further the example of positive corellation using linear regression

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "b").set(title = "Loudness & Energy")
```

```python

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "c").set(title = "Popularity & Acousticness")
```

```python
#Changing the date and column name

tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
# Creating a distribution plot
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = "hist").set(title = "Number of songs per year")
```

```python
#Duration of songs by year

total_dr = tracks.duration
fig_dims = (18,7)
fig, ax  = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year & Duration")
plt.xticks(rotation = 90)
```

```python
# Duration of songs by genre

total_dr = tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10, 5)
fig, ax  = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Year & Duration")
plt.xticks(rotation = 60)
```
